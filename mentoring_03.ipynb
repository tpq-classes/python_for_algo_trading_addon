## Features

```python
sym = 'EUR='
sym = '.SPX'
sym = 'GLD'
```

```python
data = pd.DataFrame(raw[sym])
```

```python
data.plot(figsize=(10, 6));
```

```python
data['r'] = np.log(data[sym] / data [sym].shift(1))
```

```python
window = 20
```

```python
data['risk'] = data['r'].rolling(window).std()
```

## Clustering

```python
model.fit(data[['trend', 'risk']])
```

```python
data['regime'] = model.predict(data[['trend', 'risk']])
```

## Normative Finance

### The Market

```python
S = np.array((20, 5))
B = np.array((11, 11))
M = np.array((S, B)).T
S0 = 10
B0 = 10
M0 = np.array((S0, B0))
```

```python
M0
```

```python
M
```

### The EUT Agent

```python
def u(x):
    return np.sqrt(x)
```

```python
P = np.array((0.55, 0.45))
P = np.array((0.45, 0.55))
```

```python
def EUT(x, P):
    return np.dot(u(x), P)
```

```python
x = np.array((3, 6))
```

```python
u(x)
```

```python
u(x).mean()  # for P = np.array((0.5, 0.5)) = equal probability
```

```python
EUT(x, P)  # for P as defined before
```

```python
def EUT_(phi, P):
    x = np.dot(M, phi)
    return EUT(x, P)
```

```python
phi = np.array((0.3, 0.7))
```

```python
EUT_(phi, P)
```

```python
x = np.dot(M, phi)
```

```python
x
```

```python
EUT(x, P)
```

### Optimization

```python
from scipy.optimize import minimize
```

```python
w = 10  # budget
```

```python
cons = {'type': 'eq', 'fun': lambda phi: np.dot(M0, phi) - w}   # budget constraint
```

```python
bnds = ((0, np.inf), (0, np.inf))
```

```python
opt = minimize(lambda phi, P: -EUT_(phi, P), phi, args=(P), constraints=cons, bounds=bnds)
```

```python
P
```

```python
opt
```

```python
EUT_(opt['x'], P)
```
