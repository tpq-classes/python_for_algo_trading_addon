## Finance with Python

```python
S = np.array((20, 5))
S0 = 10
```

```python
B = np.array((11, 11))
B0 = 10
```

```python
M0 = np.array((S0, B0))
M0
```

```python
M = np.array((S, B)).T
M
```

```python
K = 15
```

```python
C = np.maximum(S - K, 0)
C
```

```python
phi = np.linalg.solve(M, C)
phi
```

```python
np.dot(M, phi)
```

```python
np.dot(M, phi) == C
```

```python
np.allclose(np.dot(M, phi), C)
```

```python
C0 = np.dot(M0, phi)
C0
```
