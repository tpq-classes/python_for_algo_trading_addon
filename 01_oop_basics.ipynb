## Object Orientation

### Basics of Python Classes

This is the **syntax** to define a new (custom) class in Python.

```python
class ExampleOne(object):
    pass
```

**Instantiation** has function call syntax.

```python
c = ExampleOne()
```

```python
c
```

There are multiple **special attributes** attached to every instance of the new class.

```python
c.__str__()
```

Indeed, we have a new **object type**.

```python
type(c)
```

A **special method** is `__init__` which is called on every instantiation.

```python
class ExampleTwo(object):
    def __init__(self, a, b):
        self.a = a
        self.b = b
```

In this case, two **parameters** need to be passed.

```python
c = ExampleTwo(1, 'text')
```

**Attributes** can be looked up via the `dot` operator.

```python
c.a
```

```python
c.b
```

In general, attributes can be **overwritten**.

```python
c.a = 100
```

```python
c.a
```

Instances of classes can be (mis-) used to store arbitrary data via **data attributes**.

```python
c = ExampleOne()
```

```python
c.first_name = 'Jason'
c.last_name = 'Bourne'
c.movies = 5
```

```python
print(c.first_name, c.last_name, c.movies)
```

Functions owned by classes are called **methods**.

```python
class ExampleThree(object):
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def addition(self):
        return self.a + self.b
```

**Calling** a method.

```python
c = ExampleThree(10, 15)
```

```python
c.addition()
```

```python
c.a += 10
c.addition()
```

### Inheritance

Basic **inheritance**.

```python
class ExampleFour(ExampleTwo):
    def addition(self):
        return self.a + self.b
```

```python
c = ExampleFour(10, 15)
```

```python
c.addition()
```

**Complex** inheritance.

```python
class ExampleFive(ExampleFour):
    def multiplication(self):
        return self.a * self.b
```

```python
c = ExampleFive(10, 15)
```

```python
c.addition()
```

```python
c.multiplication()
```

Methods can be defined via **functions outside of the class definition**.

```python
def multiplication(self):
    return self.a * self.b
```

```python
class ExampleSix(ExampleFour):
    multiplication = multiplication
```

```python
c = ExampleSix(10, 15)
```

```python
c.addition()
```

```python
c.multiplication()
```

Be aware of **static definitions**.

```python
class ExampleSeven(object):
    def __init__(self, a, b):
        self.a = a
        self.b = b
        self.__sum = a + b 
        
    multiplication = multiplication
    
    def addition(self):
        return self.__sum
```

```python
c = ExampleSeven(10, 15)
```

```python
c.addition()
```

**Hierarchical access** to (private) attributes.

```python
c._ExampleSeven__sum
```

In this case, **no updating**.

```python
c.a += 10
c.a
```

```python
c.a + c.b
```

```python
c.addition()
```

```python
c._ExampleSeven__sum
```

Multiplication works as desired.

```python
c.multiplication()
```

### Encapsulation

Encapsulation basically means that data (stored as attributes of a class/object) can only be manipulated by class methods.

```python
class ExampleEight(object):
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def get_a(self):
        return self.a
    
    def get_b(self):
        return self.b
    
    def addition(self):
        return self.a + self.b
```

Adding such "getter" methods avoids directly accessing attributes.

```python
c = ExampleEight(10, 12)
```

```python
c.get_a()
```

```python
c.get_b()
```

```python
c.b
```

Changing the values of the attributes then should be managed by some "setter" methods.

```python
class ExampleNine(ExampleEight):
    def set_a(self, value):
        self.a = value
    
    def set_b(self, value):
        self.b = value
```

```python
c = ExampleNine(5, 6)
```

```python
c.addition()
```

```python
c.set_a(100)
```

```python
c.addition()
```

```python
# revisited
class ExampleSevenEnc(object):
    def __init__(self, a, b):
        self.a = a
        self.b = b
        self.__sum = a + b 
        
    multiplication = multiplication
    
    def set_a(self, value):
        self.a = value
        self.__sum = self.a + self.b
        
    def set_b(self, value):
        self.b = value
        self.__sum = self.a + self.b
    
    def addition(self):
        return self.__sum
```

```python
c = ExampleSevenEnc(10, 15)
```

```python
c.addition()
```

```python
c.set_a(20)
```

```python
c.addition()
```

### Composition

Composition refers to the fact that one object might serve as an attribute of another object.

```python
class NumberPair(object):
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def get_a(self):
        return self.a
    
    def get_b(self):
        return self.b
        
    def set_a(self, value):
        self.a = value
    
    def set_b(self, value):
        self.b = value
```

```python
class ExampleTen(object):
    def __init__(self, pair):
        self.pair = pair
        
    def addition(self):
        return self.pair.get_a() + self.pair.get_b()
    
    def multiplication(self):
        return self.pair.get_a() * self.pair.get_b()
```

```python
pair = NumberPair(10, 20)
```

```python
pair.get_a()
```

```python
c = ExampleTen(pair)
```

```python
c.addition()
```

```python
c.multiplication()
```

```python
c.pair.set_a(30)
```

```python
c.addition()
```

```python
class ExampleEleven(ExampleTen):
    def update_a(self, value):
        self.pair.set_a(value)
        
    def update_b(self, value):
        self.pair.set_b(value)
```

```python
c = ExampleEleven(pair)
```

```python
c.addition()
```

```python
c.update_a(10)
```

```python
c.addition()
```

## Python Data (or Object) Model

The so-called **Python Data or Object Model** (cf. https://docs.python.org/3/reference/datamodel.html) allows to design classes that consistently interact with basic language constructs of Python. Among others it supports (cf. Luciano Ramalho (2015): _Fluent Python_. O'Reilly, Sebastopol, p. 4):

* iteration
* collections
* attribute access
* operator overloading
* function and method invocation
* object creation and destruction
* string representation
* managed context (`with` blocks)

Following an example in _Fluent Python_ let us define a class called `Vector`.

```python
class Vector(object):
    def __init__(self, x=0, y=0, z=0):
        self.x = x
        self.y = y
        self.z = z
        
    def __repr__(self):
        return 'Vector(%r, %r, %r)' % (self.x, self.y, self.z)
```

```python
v = Vector(1, 2, 3)
```

```python
v
```

```python
print(v)
```

Adding special methods for the absolute value and boolean values.

```python
class Vector(Vector):
    def __abs__(self):
        return (self.x ** 2 +  self.y ** 2 + self.z ** 2) ** 0.5
    
    def __bool__(self):
        return bool(abs(self))
```

```python
v = Vector(1, 2, -1)
```

```python
abs(v)
```

```python
bool(v)
```

```python
v = Vector()
```

```python
v
```

```python
abs(v)
```

```python
bool(v)
```

Adding `__add__` and `__mul__` to the mix.

```python
class Vector(Vector):
    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        z = self.z + other.z
        return Vector(x, y, z)
    
    def __mul__(self, scalar):
        return Vector(self.x * scalar,
                      self.y * scalar,
                      self.z * scalar)
```

```python
v = Vector(1, 2, 3)
```

```python
v + Vector(2, 3, 4)
```

```python
v * 2
```

Adding some more ...

```python
l = ['a', 'b', '2', 5, 2.5]
```

```python
len(l)
```

```python
class Vector(Vector):
    def __len__(self):
        return 3
    
    def __getitem__(self, i):
        if i == 0: return self.x
        elif i == 1: return self.y
        elif i in [2, -1]: return self.z
        else: print('Index out of range.')
```

```python
v = Vector(1, 2, 3)
```

```python
len(v)
```

```python
v[2]
```

```python
v[3]
```

```python
v[-1]
```

Finally, we add `__iter__` for iteration.

```python
l
```

```python
for i in range(len(l)):
    print(i, l[i])
```

```python
for elem in l:
    print(elem)
```

```python
for i, elem in enumerate(l):
    print(i, elem)
```

```python
class Vector(Vector):
    def __iter__(self):
        for i in range(len(self)):
            yield self[i]
```

```python
v = Vector(1, 2, 3)
```

```python
for i in range(3):
    print(v[i])
```

```python
for coordinate in v:
    print(coordinate)
```
