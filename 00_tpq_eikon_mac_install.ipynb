## Download Workspace and Create the App Key

You can register for a **free Eikon trial** under:

https://refini.tv/3bvPev5

As a first step, you should **download and install** the Workspace application for Mac OSX from the Refinitiv download page:

https://workspace.refinitiv.com/

Once it is installed and running with you account credentials, you should be able to connect to the API via a an application key.

In Workspace you can create a **application key** for use with the locally running Workspace application via the **`APP KEY GENERATOR`** application.

You should save a text file with **filename** `eikon.cfg` having the following contents:

    [eikon]
    app_key = YOUR_APP_KEY_HERE
    
This file should be readily availble (e.g. in the current working directory) for the next steps.

## Connecting to Eikon Data API

This code sets the `app_key` to connect to the **Eikon Data API** via Workspace which needs to be running locally. It requires the previously created text file `eikon.cfg` to be in the current working directory.

```python
cfg = cp.ConfigParser()
cfg.read('eikon.cfg')  # adjust for different file location
```

```python
ek.set_app_key(cfg['eikon']['app_key'])
```

### Retrieving Historical Data

As a test, consider the retrieval of **historical end-of-day data** for the Apple, Inc. stock.

```python
data = ek.get_timeseries(['.SPX', 'AAPL.O'],  # the RICs
                         fields='*',  # all fields
                         start_date='2019-01-01',  # start date
                         end_date='2020-06-03')  # end date
```

```python
data.tail()  # final five rows
```

```python
data.info()  # DataFrame meta information
```

The code that follows visualizes the **time series data** in the `CLOSE` column as a **line plot**.

```python
cf.set_config_file(offline=True)  # set the plotting mode to offline
```

```python
data.xs('CLOSE', level=1, axis=1).tail()
```

```python
data.xs('CLOSE', level=1, axis=1).normalize().iplot()
```

### Retrieving News Headlines

We next retrieve **news headlines** for Apple, Inc.

```python
ek.get_news_headlines('R:AAPL.O macbook',  # the RIC for Apple, Inc.
                      date_from='2020-05-01T20:00:00',  # the starting time
                      date_to='2020-06-03T22:00:00',  # the end time
                      count=50)  # number of headlines
```
