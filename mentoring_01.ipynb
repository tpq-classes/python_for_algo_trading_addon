## Market

```python
S = np.array((20, 5))
```

```python
S0 = 10
```

```python
B = np.array((11, 11))
```

```python
B0 = 10
```

```python
M0 = np.array((S0, B0))
M0
```

```python
M = np.array((S, B)).T
M
```

## Option

```python
K = 8
```

```python
K - S
```

```python
P = np.maximum(K - S, 0)
P
```

## Replication

```python
np.set_printoptions(suppress=True)
```

```python
phi = np.linalg.solve(M, P)
phi
```

```python
np.dot(M, phi)
```

```python
P0 = np.dot(M0, phi)
P0
```

## Approximation

```python
phi = np.linalg.lstsq(M, P, rcond=-1)[0]
phi
```

```python
P0 = np.dot(M0, phi)
P0
```

## Risk-Neutral Pricing

```python
r = B[0] / B0 - 1
r
```

```python
def E(P_):
    return np.dot(S, P_) / (1 + r)
```

```python
P_ = np.array((0.5, 0.5))  # probability measure
```

```python
E(P_)
```

```python
E(P_) - S0
```

```python
def OF(P_):
    return abs(E(P_) - S0)
```

```python
OF(P_)
```

```python
P  # put option payoff
```

```python
from scipy.optimize import minimize
```

```python
bnds = ((0, 1), (0, 1))
```

```python
cons = ({'type': 'eq', 'fun': lambda P: P.sum() - 1})
```

```python
opt = minimize(OF, P, bounds=bnds, constraints=cons)
```

```python
opt
```

```python
Q = opt['x']  # martingale measure
Q
```

```python
OF(Q)
```

```python
P0 = np.dot(P, Q) / (1 + r)
P0
```
