```python
class StreamOanda(tpqoa.tpqoa):
    def __init__(self, conf_file):
        super(StreamOanda, self).__init__(conf_file)
        self.times = list()
        self.bids = list()
        self.asks = list()
        self.mids = list()
    def on_success(self, time, bid, ask):
        print(time, bid, ask)
        self.times.append(pd.Timestamp(time))
        self.bids.append(float(bid))
        self.asks.append(float(ask))
        self.mids.append((float(bid) + float(ask)) / 2)
        fig.data[0].x = self.times
        fig.data[0].y = self.mids
```

```python
so = StreamOanda('../../../data/pyalgo.cfg')
```

```python
so.stream_data(symbol, stop=10)
```

## Three Streams

```python
fig = go.FigureWidget()
fig.add_scatter(name=symbol)
fig.add_scatter(name='SMA1', line=dict(width=1, dash='dot'),
                mode='lines+markers')
fig.add_scatter(name='SMA2', line=dict(width=1, dash='dash'),
                mode='lines+markers')
fig
```

```python
class StreamOanda(tpqoa.tpqoa):
    def __init__(self, conf_file):
        super(StreamOanda, self).__init__(conf_file)
        self.data = pd.DataFrame()
    def on_success(self, time, bid, ask):
        print(time, bid, ask)
        self.data = self.data.append(
            pd.DataFrame({'bid': float(bid), 'ask': float(ask),
                         'mid': (float(bid) + float(ask)) / 2},
                        index=[pd.Timestamp(time)]))
        self.data['SMA1'] = self.data['mid'].rolling(5).mean()
        self.data['SMA2'] = self.data['mid'].rolling(10).mean()
        fig.data[0].x = self.data.index
        fig.data[1].x = self.data.index
        fig.data[2].x = self.data.index
        fig.data[0].y = self.data['mid']
        fig.data[1].y = self.data['SMA1']
        fig.data[2].y = self.data['SMA2']
```

```python
so = StreamOanda('../../../data/pyalgo.cfg')
```

```python
so.stream_data(symbol, stop=20)
```

## Sub-Plots

```python
from plotly.subplots import make_subplots
```

```python
f = make_subplots(rows=3, cols=1, shared_xaxes=True)
f.append_trace(go.Scatter(name=symbol), row=1, col=1)
f.append_trace(go.Scatter(name='RETURN', line=dict(width=1, dash='dot'),
                mode='lines+markers', marker={'symbol': 'triangle-up'}),
                row=2, col=1)
f.append_trace(go.Scatter(name='MOMENTUM', line=dict(width=1, dash='dash'),
                mode='lines+markers', marker={'symbol': 'x'}), row=3, col=1)
```

```python
fig = go.FigureWidget(f)
```

```python
import numpy as np
```

```python
class StreamOanda(tpqoa.tpqoa):
    def __init__(self, conf_file):
        super(StreamOanda, self).__init__(conf_file)
        self.data = pd.DataFrame()
    def on_success(self, time, bid, ask):
        print(time, bid, ask)
        self.data = self.data.append(
            pd.DataFrame({'bid': float(bid), 'ask': float(ask),
                         'mid': (float(bid) + float(ask)) / 2},
                        index=[pd.Timestamp(time)]))
        self.data['RET'] = np.log(self.data['mid'] / self.data['mid'].shift(1))
        self.data['MOM'] = self.data['RET'].rolling(10).mean()
        fig.data[0].x = self.data.index
        fig.data[1].x = self.data.index
        fig.data[2].x = self.data.index
        fig.data[0].y = self.data['mid']
        fig.data[1].y = self.data['RET']
        fig.data[2].y = self.data['MOM']
```

```python
so = StreamOanda('../../../data/pyalgo.cfg')
```
